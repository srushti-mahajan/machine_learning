Hypothesis Testing

```python
import math
from scipy.stats import norm
```

```python
#Given Data
sample_mean = 3050
X_mean = 600
sigma_X = 25
n = 25
```

```python
#Theoratical mean cost (μ0)
theoretical_mean = 1000 + 5 * X_mean
sigma_W = 5 * sigma_X
```

```python
theoretical_mean
```

```python
sigma_W
```

```python
#Calculate test statistic t test
t=(sample_mean - theoretical_mean) / (sigma_W / math.sqrt(n))
```

```python
t
```

```python
#Using the alpha level of 5% (α = 0.05), determine the critical value from the standard normal (Z) distribution table.
alpha=0.05
critical_value=norm.ppf(1-alpha)
```

```python
critical_value
```

```python
#print all result
print(f"Theoretical Mean Cost (μ0): {theoretical_mean}")
print(f"Test Statistic (t): {t:.2f}")
print(f"Critical Value (Z): {critical_value:.2f}")
```

```python
#Make a Decision:
if t > critical_value:
    print("Reject the null hypothesis. The actual weekly operating cost is higher than the theoretical model.")


else:
     print("Fail to reject the null hypothesis.There is not enough evidence to support the claim that the actual weekly operating costs are higher.")
```
